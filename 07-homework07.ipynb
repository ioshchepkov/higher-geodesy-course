# Домашнее задание №7: Вычисление координат с помощью ИСЗ относительным методом

При относительных спутниковых определениях находят приращения координат определяемого пункта сети относительно исходного по измерениям расстояний от ИСЗ до обоих пунктов. Принцип
относительных определений пояснен на рисунке ниже.

![ ](https://zenodo.org/api/iiif/record:10868847:fig_hw7.png/full/!500,500/0/default.png "Схема относительных спутниковых определений координат")

На рисунке: 1 и 2 $-$ опорный и определяемый пункты; ИСЗ $-$ искусственный спутник Земли; $О$ $-$ начало прямоугольной геоцентрической системы координат $X, Y, Z$; $\bar{\rho}_1, \bar{\rho}_2 -$ векторы "спутник-приёмник"; $\bar{r}_1, \bar{r}_2, \bar{r}_i -$ геоцентрические векторы геодезических пунктов и ИСЗ; $\bar{D} - $ вектор расстояния от базовой станции до определяемой.

Вектор $\bar{D}_{12}$ равен как разности геоцентрических радиусов векторов станций, так и разности векторов «спутник-приемник»:

$$
\bar{D}_{12} = \bar{r}_2 - \bar{r}_1 = \bar{\rho}_2 - \bar{\rho}_1
$$

Найдем разность квадратов векторов $\bar{\rho}_1$ , $\bar{\rho}_2$, рассматривая квадрат вектора как скалярное произведение:

$$
\bar{\rho}_2^2 - \bar{\rho}_1^2 = \rho_2^2 - \rho_1^2 = (\rho_2 - \rho_1)(\rho_2 + \rho_1) = 2(\rho_2 - \rho_1)\rho_{\text{ср}},
$$

где $\rho_{\text{ср}} - $ среднее арифметическое из расстояний "спутник-приемник" и вычисляется по формуле:

$$
\rho_{\text{ср}} = \frac{\rho_2 + \rho_1}{2}
$$

Теперь разность $\bar{\rho}_2^2 - \bar{\rho}_1^2$ выразим через геоцентрические радиусы-векторы.

$$
\bar{\rho}_1 = \bar{r}_1 - \bar{r}_i
$$

$$
\bar{\rho}_2 = \bar{r}_2 - \bar{r}_i
$$

Для разности квадратов векторов $\bar{\rho}_2$ и $\bar{\rho}_1$ получаем

$$
\bar{\rho}_2^2 - \bar{\rho}_1^2 = (\bar{r}_2 - \bar{r}_i)^2 - (\bar{r}_1 - \bar{r}_i)^2 = \bar{r}_2^2 - \bar{r}_1^2 - 2(\bar{r}_2 - \bar{r}_1)\bar{r}_i = 2(\bar{r}_2 - \bar{r}_1)(\bar{r}_{\text{ср}} - \bar{r}_i) = 2\bar{D}(\bar{r}_{\text{cp}} - \bar{r}_i),
$$

где $\bar{r}_{\text{cp}} - $ среднее из радиусов-векторов пунктов 1 и 2 и вычисляется по формуле:

$$
\bar{r}_{\text{cp}} = \frac{\bar{r}_2 - \bar{r}_1}{2}
$$

Приравняем выражения для $\bar{\rho}_2^2 - \bar{\rho}_1^2$:

$$
(\rho_2 - \rho_1)\rho_{\text{ср}} = \bar{D}(\bar{r}_{\text{cp}} - \bar{r}_i)
$$

В правой части этого выражения стоит скалярное произведение векторов. Для его вычисления запишем в развернутом виде векторы $\bar{D}, \bar{r}_{\text{cp}}, \bar{r}_i$:

$$
\bar{D} = (X_2 - X_1)i + (Y_2 -Y_1)j + (Z_2 - Z_1)k,
$$

$$
\bar{r}_{\text{cp}} = X_{\text{cp}}i + Y_{\text{cp}}j + Z_{\text{cp}}k
$$

$$
\bar{r}_i = X_{i}i + Y_{i}j + Z_{i}k
$$

$$
\bar{r}_{\text{cp}} - \bar{r}_i = (X_{\text{cp}} - X_{i})i + (Y_{\text{cp}} - Y_{i})j + (Z_{\text{cp}} - Z_{i})k,
$$

где $i, j, k - $ единичные векторы, $X_{\text{cp}}, Y_{\text{cp}}, Z_{\text{cp}} - $ среднее из координат пунктов 1 и 2, $X_i, Y_i, Z_i - $ координаты спутника.  

Получаем скалярное произведение:

$$\tag{1}
\bar{D}(\bar{r}_{\text{ср}} - \bar{r}_i) = (X_{\text{ср}} - X_i)(X_2 - X_1) + (Y_{\text{ср}} - Y_i)(Y_2 - Y_1) + (Z_{\text{ср}} - Z_i)(Z_2 - Z_1) = (\rho_2 - \rho_1)\rho_{\text{ср}},
$$

где $\rho_{\text{ср}} - $ расстояние между ИСЗ и точкой с координатами $X_{\text{ср}}, Y_{\text{ср}}, Z_{\text{ср}}$ высичляется по формуле:

$$
\rho_{\text{ср}} = \sqrt{(X_{\text{ср}} - X_i)^2 + (Y_{\text{ср}} - Y_i)^2 + (Z_{\text{ср}} - Z_i)^2}
$$

Разделим уравнение (1) на $\rho_{\text{ср}}$:

$$\tag{2}
(\rho_2 - \rho_1)_i = (X_2 - X_1)\frac{X_{\text{ср}} - X_i}{\rho_{\text{ср}}} + (Y_2 - Y_1)\frac{Y_{\text{ср}} - Y_i}{\rho_{\text{ср}}} + (Z_2 - Z_1)\frac{Z_{\text{ср}} - Z_i}{\rho_{\text{ср}}}
$$

Введём обозначения для направлющих косинусов линии $\rho_{\text{ср}}$:

$$
l_i = \frac{X_{\text{ср}} - X_i}{\rho_{\text{ср}}}
$$

$$
m_i = \frac{Y_{\text{ср}} - Y_i}{\rho_{\text{ср}}}
$$

$$
n_i = \frac{Z_{\text{ср}} - Z_i}{\rho_{\text{ср}}}
$$

В таком случае уравнение (2) можно представить в виде:

$$\tag{3}
(\rho_2 - \rho_1)_i = (X_2 - X_1)l_i + (Y_2 - Y_1)m_i + (Z_2 - Z_1)n_i, 
$$

здесь разность $(\rho_2 - \rho_1)_i - $ является результатом измерений при спутниковых наблюдениях; приращения координат $(X_2 - X_1), (Y_2 - Y_1), (Z_2 - Z_1) - $ определяемые величины; направляющие косинусы $l_i, m_i, n_i - $ коэффициенты уравнений.

В уравнении (3) три неизвестных величины, поэтому для определения приращения координат необходимо одновременно наблюдать минимум 3 спутника. В таком случае будем иметь систему из трёх уравнений с тремя определяемыми величинами:

$$
\begin{matrix}
(\rho_2 - \rho_1)_i = (X_2 - X_1)l_i + (Y_2 - Y_1)m_i + (Z_2 - Z_1)n_i\\
(\rho_2 - \rho_1)_j = (X_2 - X_1)l_j + (Y_2 - Y_1)m_j + (Z_2 - Z_1)n_j\\
(\rho_2 - \rho_1)_k = (X_2 - X_1)l_k + (Y_2 - Y_1)m_k + (Z_2 - Z_1)n_k
\end{matrix}
$$

Или в матричном виде:

$$
\begin{bmatrix}
(\rho_2 - \rho_1)_i\\
(\rho_2 - \rho_1)_j\\
(\rho_2 - \rho_1)_k
\end{bmatrix}
=
\begin{bmatrix}
l_i & m_i & n_i\\
l_j & m_j & n_j\\
l_k & m_k & n_k
\end{bmatrix}
\begin{bmatrix}
(X_2 - X_1)\\
(Y_2 - Y_1)\\
(Z_2 - Z_1)
\end{bmatrix}
$$

Решение данной системы выполняется по формуле:

$$
\begin{bmatrix}
(X_2 - X_1)\\
(Y_2 - Y_1)\\
(Z_2 - Z_1)
\end{bmatrix}
=
\begin{bmatrix}
l_i & m_i & n_i\\
l_j & m_j & n_j\\
l_k & m_k & n_k
\end{bmatrix}^{-1}
\begin{bmatrix}
(\rho_2 - \rho_1)_i\\
(\rho_2 - \rho_1)_j\\
(\rho_2 - \rho_1)_k
\end{bmatrix}
$$
