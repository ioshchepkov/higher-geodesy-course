Задание посвящено определению координат пункта методом обратной линейной пространственной засечки.

Разберём каждое слово отдельно. 

**Засечка** означает, что определение координат выполняется путём выполнения измерений некоторых величин между опорным и определяемым пунктом.

**Обратная** означает, что средство наблюдений (геодезический прибор) установлен на определяемом пункте. Бывает *прямая засечка*; в этом случае средство наблюдений устанавливается на опорный пункт.

**Пространственная** означает, что будут определяться простанственные координаты. 

**Линейная** означает, что измеряемой величиной является расстояние между опорным и определяемым пунктом. Бывают *угловые засечки*; в этом случае измеренными величинами являются углы между направлениями на опорные станции. А также бывают *линейно-угловые засечки*; в этом случае измеренными величинами являются как углы, так и расстояния.


## Обратная линейная пространственная засечка

В современном мире, для установления глобальных систем отсчёта координат, используются методы космической геодезии. В этом случае, в качестве опорных пунктов зачастую выступают искусственные спутники Земли. Для определения координат определяемого пункта, необходимо знать расстояние минимум до трёх опорных. Схема обратной линейной пространственной засечки представлена на рисунке ниже.

![ ](https://zenodo.org/api/iiif/record:10829269:fig1.png/full/!800,800/0/default.png "Линейная пространственная засечка")


На рисунке $P - $ определяемый пункт; $X_0, Y_0, Z_0 - $ его координаты, которые следует найти; $1, 2, 3 - $ спутники (опорные пункты); $P_1, P_2, P_3 - $ измеренные расстояния между опорными и определяемым пунктом.

Искомые координаты определяют по формулам:

\begin{equation}\tag{1}
\begin{matrix}
X_P = X_i + P_il_{i0} \\
Y_P = Y_i + P_im_{i0} \\
Z_P = Z_i + P_in_{i0}
\end{matrix},
\end{equation}

где $X_i, Y_i, Z_i - $ координаты любого опорного пункта; $P_i - $ расстояние от него до определяемого, $l_{i0}, m_{i0}, n_{i0} - $ направляющие косинусы линии $P_i$. Следовательно, для нахождения координат точки $Р$ нужно найти направляющие косинусы любого из векторов $Р_i$. 

**Напоминание.** Косинусом угла называется отношение прилежащего катета к гипотенузе. 

**Напоминание.** Направляющие косинусы вектора (в пространстве) - это косинусы углов, которые вектор образует с положительными полуосями координат.

Найдем направляющие косинусы $l_{10}$ , $m_{10}$ , $n_{10}$ вектора $Р_1$. Воспользуемся для этого формулой, определяющей косинус угла между векторами
$Р_1$ и $D_{1-2}$ или $D_{1-3}$ через направляющие косинусы этих векторов.

Для векторов $D_{1-2}$ и $Р_1$:

$$
\cos\alpha_{P_{12}} = L_{12}l_{10} + M_{12}m_{10} + N_{12}n_{10},
$$

для векторов $D_{1-3}$ и $Р_1$:

$$
\cos\alpha_{P_{13}} = L_{13}l_{10} + M_{13}m_{10} + N_{13}n_{10},
$$

$L_{12}, M_{12}, N_{12}, L_{13}, M_{13}, N_{13} - $ направляющие косинусы векторов $D_{1-2}$ и $D_{1-3}$ соответственно. Они вычисляются по формуле:

$$
\begin{matrix}
L_{1i} = \frac{X_i - X_1}{D_{1i}}, & M_{1i} = \frac{Y_i - Y_1}{D_{1i}}, & N_{1i} = \frac{Z_i - Z_1}{D_{1i}}, & i = 2, 3.
\end{matrix}
$$

В двух уравнениях для $\cos\alpha_{P_{12}}$ и $\cos\alpha_{P_{13}}$ содержатся три неизвестных $l_{10}, m_{10}, n_{10}.$ Чтобы их определить, к уравнениям для $\cos\alpha_{P_{12}}$ и $\cos\alpha_{P_{13}}$ добавляют тождество:

$$
l_{10}^2 + m_{10}^2 + n_{10}^2 = 1
$$

Таким образом, получается система из трёх уравнений с тремя неизвестными:

\begin{equation}\tag{2}
 \begin{cases}
     \cos\alpha_{P_{12}} = L_{12}l_{10} + M_{12}m_{10} + N_{12}n_{10} \\
     \cos\alpha_{P_{13}} = L_{13}l_{10} + M_{13}m_{10} + N_{13}n_{10} \\
     1 = l_{10}^2 + m_{10}^2 + n_{10}^2
 \end{cases}
\end{equation}


Значения углов $\cos\alpha_{P_{12}}$ и $\cos\alpha_{P_{13}}$ можно найти из теоремы косинусов для треугольников $1P2$ и $1P3$, соответственно.

Запишем теорему косинусов для треугольника $1P2$

$$
P_2^2 = P_1^2 + D_{1-2}^2 - 2P_1D_{1-2}\cdot\cos\alpha_{P_{12}},
$$

Тогда 

$$
\cos\alpha_{P_{12}} = \frac{P_1^2 + D_{1-2}^2 - P_2^2}{2P_1D_{1-2}},
$$

где $P_i - $ измеренные расстояния; $D_{1-2} - $ расстояние между спутниками 1 и 2, вычисляемое по формуле:

$$
D_{1-2} = \sqrt{(X_2 - X_1)^2 + (Y_2 - Y_1)^2 + (Z_2 - Z_1)^2}
$$

Рассмотрим одну из возможных методик решения системы (2). В первых двух уравнениях данной системы выразим $l_{10}$ и $m_{10}$ через $n_{10}$:

$$
\cos\alpha_{P_{12}} - N_{12}n_{10} = L_{12}l_{10} + M_{12}m_{10}
$$
$$
\cos\alpha_{P_{13}} - N_{13}n_{10} = L_{13}l_{10} + M_{13}m_{10}
$$

или в матричном виде:

$$
\begin{pmatrix}
\cos\alpha_{P_{12}} - N_{12}n_{10} \\
\cos\alpha_{P_{13}} - N_{13}n_{10}
\end{pmatrix}
=
\begin{pmatrix}
L_{12} &  M_{12}\\
L_{13} &  M_{13}\\
\end{pmatrix}
\begin{pmatrix}
l_{10}\\
m_{10}\\
\end{pmatrix}.
$$

Решив это уравнение, выразим направляющие косинусы $l_{10}, m_{10}$ через $n_{10}$:

$$
\begin{pmatrix}
l_{10}\\
m_{10}\\
\end{pmatrix}
=
\begin{pmatrix}
L_{12} &  M_{12}\\
L_{13} &  M_{13}\\
\end{pmatrix}^{-1}
\begin{pmatrix}
\cos\alpha_{P_{12}} - N_{12}n_{10} \\
\cos\alpha_{P_{13}} - N_{13}n_{10}
\end{pmatrix}
$$


В правой части разделим известные и неизвестные величины

$$
\begin{pmatrix}
l_{10}\\
m_{10}\\
\end{pmatrix}
=
\begin{pmatrix}
L_{12} &  M_{12}\\
L_{13} &  M_{13}\\
\end{pmatrix}^{-1}
\begin{pmatrix}
\cos\alpha_{P_{12}}\\
\cos\alpha_{P_{13}}\\
\end{pmatrix}
-
\begin{pmatrix}
L_{12} &  M_{12}\\
L_{13} &  M_{13}\\
\end{pmatrix}^{-1}
\begin{pmatrix}
N_{12}\\
N_{13}
\end{pmatrix}
n_{10}
$$


Первую матрицу правой части обозначим через $A$:

$$
A = 
\begin{pmatrix}
L_{12} &  M_{12}\\
L_{13} &  M_{13}\\
\end{pmatrix}^{-1}
\begin{pmatrix}
\cos\alpha_{P_{12}}\\
\cos\alpha_{P_{13}}\\
\end{pmatrix}
=
\begin{pmatrix}
A_1\\
A_2
\end{pmatrix}
$$

известную часть второй - через $В$:

$$
B = 
\begin{pmatrix}
L_{12} &  M_{12}\\
L_{13} &  M_{13}\\
\end{pmatrix}^{-1}
\begin{pmatrix}
N_{12}\\
N_{13}
\end{pmatrix}
=
\begin{pmatrix}
B_1\\
B_2
\end{pmatrix}
$$

и напишем

$$
\begin{pmatrix}
l_{10}\\
m_{10}
\end{pmatrix}
=
\begin{pmatrix}
A_1 & B_1\\
A_3 & B_3
\end{pmatrix}
\begin{pmatrix}
1\\
n_{10}
\end{pmatrix}
=
\begin{pmatrix}
A_1 + B_1n_{10}\\
A_3 + B_3n_{10}
\end{pmatrix}
$$

В результате сделанных преобразований из исходной системы (2) получаем такую систему:

\begin{equation}\tag{3}
 \begin{cases}
     l_{10} = A_1 +  B_1n_{10}\\
     m_{10} = A_3 + B_3n_{10} \\
     n_{10}^2 = 1 - l_{10}^2 - m_{10}^2
 \end{cases}
\end{equation}

Подставив в третье уравнение этой системы выражения для $l_{10}$ и $m_{10}$ из первых двух, получаем квадратное уравнение для $n_{10}$:

$$
(1 + B_1^2 + B_3^2)n_{10}^2 + 2(A_1B_1 + A_3B_3)n_{10} + (A_1^2 + A_3^2 - 1) = 0
$$

и для для северного полушария $n_{10}$ находим по формуле:

$$
n_{10} = \frac{-(A_1B_1 + A_3B_3) - \sqrt{(A_1B_1 + A_3B_3)^2 - (1 + B_1^2 + B_3^2)(A_1^2 + A_3^2 - 1)}}{1 + B_1^2 + B_3^2}
$$

для южного полушария:

$$
n_{10} = \frac{-(A_1B_1 + A_3B_3) + \sqrt{(A_1B_1 + A_3B_3)^2 - (1 + B_1^2 + B_3^2)(A_1^2 + A_3^2 - 1)}}{1 + B_1^2 + B_3^2}
$$


После нахождения $n_{10}$ находим значения $l_{10}$ и $m_{10}$ по формулам (3). Контролем правильности вычислений служит третье уравнение из системы (3).
