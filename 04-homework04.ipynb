## Законы Кеплера
Невозмущённое движение спутников называют ещё Кеплеровым движением, так как оно подчиняется трём законам, открытым в 1609-1618 гг. немецким астрономом Иоганном Кеплером. Исходная их формулировка относилась к планетам, движущимся вокруг Солнца. Однако законы справедливы для любого движения тел в гравитационном поле других, значительно более массивных тел. Например, для движения ИСЗ в вокруг Земли.  

**Первый закон.** Каждая планета Солнечной системы движется по эллипсу вокруг Солнца, находящегося в одном из его фокусов.

## Кеплеровы элементы орбиты
Орбиту ИСЗ при невозмущённом движении можно однозначно задать в виде шести кеплеровых элементов орбиты, заданных в момент $t_0$.

К ним относятся: $a -$ большая полуось эллипса. Характеризует размер орбиты; $e -$ эксцентриситет эллипса. Характеризует форму орбиты; $\Omega - $ долгота восходящего узла; $i - $ наклон орбиты. Углы $\Omega$ и $i$ задают ориентировку орбитальной плоскости в пространстве; $\omega - $ аргумент перицентра. Задаёт ориентировку орбитального эллипса в плоскости орбиты; $\vartheta - $ истинная аномалия. Задаёт положение спутника на орбите. Вместо истинной аномалии используются также либо эксцентрическая аномалия $E$, либо средняя аномалия $M$.

### Вычисление прямоугольных координат ИСЗ

Пусть заданы элементы орбиты $a, e, i, \Omega, \omega, M_0$ в момент времени $t_0$. Требуется на заданный момент времени $t$ вычислить прямоугольные координаты $x, y, z$. 

Для решения данной задачи выполняется следующий алгоритм. 


1. Вычисляем среднее движение ИСЗ:

$$
n = \frac{\sqrt{GM}}{a\sqrt{a}},
$$
где $GM - $ геоцентрическая гравитационная постоянная 

2. Вычисляем среднюю аномалию на заданный момент времени:

$$
M = M_0 + n(t-t_0)
$$

3. Методом последовательных приближений вычисляем эксцентрическую аномалию:

$$
E_0 = M + e\sin M;
$$

$$
E_k = M + e\sin E_{k-1}
$$

Вычисления прекращаем, когда 

$$
\left| E_k - E_{k-1} \right| < 0.0000001^{\circ}
$$

4. Вычисляем истиную аномалию:

$$
\vartheta = E + 2\arctan\left(\frac{e\sin E}{1 + \sqrt{1 - e^2} - e\cos E}\right)
$$

5. Вычисляем аргумент широты:

$$
u = \omega + \vartheta
$$

6. Вычисляем геоцентрическое расстояние до спутника:

$$
r = \frac{a(1 - e^2)}{1 + e\cos\vartheta}
$$

7. Вычисляем прямоугольные координаты спутника:

$$
\begin{pmatrix}
x\\
y\\
z
\end{pmatrix}
=
r
\begin{pmatrix}
\cos\Omega\cos u - \sin\Omega\sin u \cos i\\
\sin\Omega\cos u + \cos\Omega\sin u \cos i\\
\sin u\sin i
\end{pmatrix}
$$

### Виды орбит ИСЗ
Орбиты ИСЗ классифицируются по различным признакам

1. По высоте геоцентрической орбиты

   1.1 Низкая околоземная — геоцентрическая орбита с высотой до 2000 км

   1.2 Средневысокая — геоцентрическая орбита с высотой выше 2000 км, но ниже геосинхронной орбиты (35786 км) (на этой орбите находятся спутниковые системы навигации — GPS, ГЛОНАСС и др.)
   
   1.3 Геосинхронная — геоцентрическая орбита на высоте 35786 км, на которой орбитальный период равен звёздным суткам Земли (периоду вращения Земли вокруг своей оси); частным случаем является геостационарная орбита, имеющая нулевое наклонение относительно экватора Земли

2. По эксцентриситету орбиты

   2.1 Круговая — орбита с эксцентриситетом $e = 0$, имеющая форму окружности

   2.2 Эллиптическая — орбита с эксцентриситетом $0 < e < 1$, имеющая форму эллипса

3. По наклонению орбиты

   3.1 Наклонная — орбита с наклонением $i > 0^\circ$ относительно плоскости экватора; частным случаем является полярная орбита с наклонением $i=90^\circ$.

   3.2 Экваториальная — орбита с наклонением $i = 0^\circ$ относительно экватора Земли; частными случаями являются геостационарная орбита
